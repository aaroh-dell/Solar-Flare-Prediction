# # Project: Predicting Solar Flares using Multi-Task Deep Learning
**Objective:** Build a neural network capable of processing SDO (Solar Dynamics Observatory) satellite magnetograms to predict both the exact physical magnitude ($W/m^2$) and the classification risk of imminent solar flares.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms
from PIL import Image, ExifTags
import glob
import random
import os

# 1. Load the massive dataset ONCE
metadata_path = '/kaggle/input/datasets/organizations/fhnw-i4ds/sdobenchmark/SDOBenchmark_full/training/meta_data.csv'
df = pd.read_csv(metadata_path)

# 2. Universal Physics Functions
def get_magnitude(flux):
    if flux >= 1e-4: return f"X{flux*1e4:.2f}"
    if flux >= 1e-5: return f"M{flux*1e5:.2f}"
    if flux >= 1e-6: return f"C{flux*1e6:.2f}"
    return f"B{flux*1e7:.2f}"

def get_class(flux):
    if flux >= 1e-4: return "X"
    if flux >= 1e-5: return "M"
    if flux >= 1e-6: return "C"
    return "B"

df['magnitude'] = df['peak_flux'].apply(get_magnitude)
df['flare_class'] = df['peak_flux'].apply(get_class)

print(f"✅ Foundation Ready: Loaded {len(df)} Solar Events!")
```

```python
# THE AUTO-TRASH DATASET
class SolarFlareDataset(Dataset):
    def __init__(self, dataframe, root_dir, transform=None):
        self.df = dataframe
        self.root_dir = root_dir
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        item_id, peak_flux = row['id'], row['peak_flux']
        
        parts = item_id.split('_')
        ar_folder = parts[0]
        timestamp_folder = "_".join(parts[1:])
        
        folder_path = os.path.join(self.root_dir, ar_folder, timestamp_folder)
        img_paths = glob.glob(f"{folder_path}/*magnetogram.jpg")
        if not img_paths:
            img_paths = glob.glob(f"{folder_path}/*.jpg") 
        img_path = img_paths[0]
        
        # The EXIF "Flagged" Filter
        image = Image.open(img_path)
        exif = image.getexif()
        if exif is not None:
            for tag_id, value in exif.items():
                if isinstance(value, str) and 'flagged' in value.lower():
                    return self.__getitem__((idx + 1) % len(self))
        
        image = image.convert('L')
        if self.transform:
            image = self.transform(image)
            
        return image, torch.tensor(peak_flux, dtype=torch.float32)

data_transforms = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5]) 
])

ROOT_DIR = '/kaggle/input/datasets/organizations/fhnw-i4ds/sdobenchmark/SDOBenchmark_full/training'
print("✅ Blueprint Ready: Neural Network Ready.")
```

```python
print("🔬 Preventing Data Leakage: Splitting by Active Region...")

unique_regions = df['active_region'].unique()
np.random.seed(42) 
np.random.shuffle(unique_regions)

split_idx = int(0.8 * len(unique_regions))
train_regions = unique_regions[:split_idx]
test_regions = unique_regions[split_idx:]

train_df = df[df['active_region'].isin(train_regions)].reset_index(drop=True)
test_df = df[df['active_region'].isin(test_regions)].reset_index(drop=True)

train_dataset = SolarFlareDataset(train_df, ROOT_DIR, transform=data_transforms)
test_dataset = SolarFlareDataset(test_df, ROOT_DIR, transform=data_transforms)

print("⚖️ Calculating class weights for the Smart Sampler...")
class_counts = train_df['magnitude'].apply(lambda x: x[0]).value_counts().to_dict()
class_weights = {k: 1.0 / v for k, v in class_counts.items()}
sample_weights = [class_weights[get_magnitude(row['peak_flux'])[0]] for _, row in train_df.iterrows()]

sampler = WeightedRandomSampler(
    weights=sample_weights, 
    num_samples=len(sample_weights), 
    replacement=True 
)

train_loader = DataLoader(train_dataset, batch_size=64, sampler=sampler)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

print(f"🚀 Master Pipeline Ready!")
print(f"🏋️ Gym: {len(train_dataset)} flares (from {len(train_regions)} unique sunspots)")
print(f"📝 Vault: {len(test_dataset)} flares (from {len(test_regions)} unique sunspots)")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from sklearn.metrics import classification_report
from torch.utils.data import DataLoader

print("Phase 1: Testing the 'Amnesia' Baseline (Guaranteed Failure)...")

def get_class_index_naive(flux):
    if flux >= 1e-4: return 3 # X
    if flux >= 1e-5: return 2 # M
    if flux >= 1e-6: return 1 # C
    return 0 # B

# 1. THE SABOTAGED BLUEPRINT
class AmnesiaCNN(nn.Module):
    def __init__(self):
        super(AmnesiaCNN, self).__init__()
        # SABOTAGE 1: 1x1 Kernel. It cannot see shapes, edges, or magnetic shear.
        self.conv1 = nn.Conv2d(1, 2, kernel_size=1) 
        
        # SABOTAGE 2: Massive Average Pooling. Blurs the image into gray mush.
        self.pool = nn.AvgPool2d(16, 16) 
        
        # SABOTAGE 3: Extreme Bottleneck.
        # Assuming 128x128 image -> pool(16) -> 8x8. 2 channels * 8 * 8 = 128 features total.
        self.fc1 = nn.Linear(128, 4) 

    def forward(self, x):
        # Forward pass with crippled vision
        x = self.pool(F.relu(self.conv1(x)))
        x = x.view(x.size(0), -1)
        return self.fc1(x)

naive_model = AmnesiaCNN().to('cuda')

# 2. DATA LOADER (Ensure train_dataset is used here)
naive_train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

# 3. SABOTAGED TRAINING LOOP 
criterion = nn.CrossEntropyLoss()
# SABOTAGE 4: Basic SGD with an absurdly high learning rate (1.0).
# The model will violently oscillate and learn nothing.
optimizer = optim.SGD(naive_model.parameters(), lr=1.0) 

print("Training Amnesia Model (It is physically incapable of learning)...")
naive_model.train()
for epoch in range(5):
    for inputs, targets in naive_train_loader:
        inputs = inputs.to('cuda')
        target_indices = torch.tensor([get_class_index_naive(t.item()) for t in targets], dtype=torch.long).to('cuda')
        
        optimizer.zero_grad()
        outputs = naive_model(inputs)
        loss = criterion(outputs, target_indices)
        loss.backward()
        optimizer.step()

# 4. THE TOTAL FAILURE REPORT 
print("\n--- 'AMNESIA' CNN CLASSIFICATION REPORT ---")
naive_model.eval()
all_preds, all_actuals = [], []

with torch.no_grad():
    for inputs, targets in test_loader: 
        inputs = inputs.to('cuda')
        outputs = naive_model(inputs)
        _, predicted = torch.max(outputs, 1)
        
        all_preds.extend(predicted.cpu().numpy())
        all_actuals.extend([get_class_index_naive(t.item()) for t in targets])

classes = ['B-Class', 'C-Class', 'M-Class', 'X-Class']
pred_labels = [classes[i] for i in all_preds]
actual_labels = [classes[i] for i in all_actuals]

print(classification_report(actual_labels, pred_labels, labels=classes, zero_division=0))
```

```python
import torch.nn as nn
import torchvision.models as models

class SolarResNet_MultiTask(nn.Module):
    def __init__(self):
        super(SolarResNet_MultiTask, self).__init__()
        # 1. The Main "Eyes" (Feature Extractor)
        self.backbone = models.resnet18(pretrained=True)
        self.backbone.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        
        # 2. Cut off the original final layer, let the raw 512 features pass through
        num_ftrs = self.backbone.fc.in_features
        self.backbone.fc = nn.Identity() 

        # 3. HEAD 1: The Probability Engine (4 outputs)
        self.classifier_head = nn.Linear(num_ftrs, 4)
        
        # 4. HEAD 2: The Exact Magnitude Engine (1 output)
        self.regression_head = nn.Linear(num_ftrs, 1)

    def forward(self, x):
        # Extract the deep visual features from the image
        features = self.backbone(x)
        
        # Send those features to BOTH heads simultaneously
        class_logits = self.classifier_head(features)
        flux_pred = self.regression_head(features)
        
        # Return both answers!
        return class_logits, flux_pred

model = SolarResNet_MultiTask().to('cuda')
print("🧠 Step 3 Complete: Two-Headed Multi-Task Brain Ready!")
```

```python
import torch.optim as optim
import torch
import torch.nn as nn

def get_class_index(flux):
    if flux >= 1e-4: return 3 # X
    if flux >= 1e-5: return 2 # M
    if flux >= 1e-6: return 1 # C
    return 0 # B

def train_multitask_resnet(model, loader, epochs=30):
    # Two different grading rubrics!
    class_criterion = nn.CrossEntropyLoss()
    reg_criterion = nn.MSELoss()
    
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    model.train()
    print("🛰️ Step 4: Training the Two-Headed Multi-Task Engine...")
    
    for epoch in range(epochs):
        running_loss = 0.0
        for inputs, targets in loader:
            inputs, targets = inputs.to('cuda'), targets.to('cuda')
            
            # Prepare answers for Head 1 (Bucket IDs) and Head 2 (Log10 Math)
            target_indices = torch.tensor([get_class_index(t.item()) for t in targets], dtype=torch.long).to('cuda')
            log_targets = torch.log10(targets + 1e-9)
            
            optimizer.zero_grad()
            
            # AI outputs BOTH answers
            class_logits, flux_preds = model(inputs)
            
            # Grade both heads
            loss_class = class_criterion(class_logits, target_indices)
            loss_reg = reg_criterion(flux_preds.squeeze(), log_targets)
            
            # THE SECRET: Add the penalties together!
            total_loss = loss_class + loss_reg 
            
            total_loss.backward()
            optimizer.step()
            running_loss += total_loss.item()
            
        print(f"Epoch {epoch+1}/{epochs} | Combined Loss: {running_loss/len(loader):.4f}")

# Train the beast!
train_multitask_resnet(model, train_loader, epochs=30)

torch.save(model.state_dict(), 'solar_multitask_final.pth')
print("\n💾 TWO-HEADED MODEL SAVED!")
```

```python
import matplotlib.pyplot as plt
import torch
import numpy as np

print("🔍 Hunting for the perfect C/M Boundary Illusion...")

model.eval()
classes = ['B', 'C', 'M', 'X']

c_class_candidates = []
m_class_candidates = []

with torch.no_grad():
    for idx in range(len(test_dataset)):
        img, target = test_dataset[idx]
        flux = target.item()
        
        if 1e-6 <= flux < 1e-5:
            c_class_candidates.append((flux, img))
        elif 1e-5 <= flux < 1e-4:
            m_class_candidates.append((flux, img))

# Sort them (Highest C first, Lowest M first)
c_class_candidates.sort(key=lambda x: x[0], reverse=True)
m_class_candidates.sort(key=lambda x: x[0])

# --- THE SMART FILTER ---
final_c = None
final_m = None

with torch.no_grad():
    # 1. Find a High C-Class where the AI confidently guessed 'C'
    for flux, img in c_class_candidates:
        logits, _ = model(img.unsqueeze(0).to('cuda'))
        pred_class = classes[torch.argmax(logits, dim=1).item()]
        if pred_class == 'C':
            final_c = (flux, img, pred_class)
            break
            
    # 2. Find a Low M-Class where the AI got confused and guessed 'C'
    for flux, img in m_class_candidates:
        logits, _ = model(img.unsqueeze(0).to('cuda'))
        pred_class = classes[torch.argmax(logits, dim=1).item()]
        if pred_class == 'C':
            final_m = (flux, img, pred_class)
            break

if final_c and final_m:
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    
    # High C-Class Image
    axes[0].imshow(final_c[1].squeeze(), cmap='gray')
    axes[0].set_title(f"Actual: {get_magnitude(final_c[0])}\nAI Guessed: {final_c[2]}-Class", fontweight='bold')
    axes[0].axis('off')
    
    # Low M-Class Image
    axes[1].imshow(final_m[1].squeeze(), cmap='gray')
    axes[1].set_title(f"Actual: {get_magnitude(final_m[0])}\nAI Guessed: {final_m[2]}-Class", fontweight='bold')
    axes[1].axis('off')
    
    plt.suptitle("The Boundary Illusion: Why the AI Struggles with M-Class", fontsize=16, fontweight='bold', y=1.05)
    plt.tight_layout()
    plt.show()
else:
    print("Could not find the perfect matching pair in this test split.")
```

```python
# --- THE NEW 3-CLASS TRANSLATOR ---
def get_class_index(flux):
    if flux >= 1e-4: return 2 # X-Class (Extreme)
    if flux >= 1e-5: return 1 # M-Class (Warning)
    return 0 # B or C Class (Safe)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import torch

model.eval()
all_preds = []
all_actuals = []

# Our new simplified mission categories
classes = ['Safe (B/C)', 'M-Class', 'X-Class']

print("🛰️ Scanning the Vault: M-Class Early Warning System...")

with torch.no_grad():
    for inputs, targets in test_loader:
        inputs = inputs.to('cuda')
        class_logits, _ = model(inputs)
        
        probabilities = torch.softmax(class_logits, dim=1)
        batch_predictions = []
        
        for prob in probabilities:
            # prob[0] = Safe, prob[1] = M, prob[2] = X
            
            # THE "MIDDLE GROUND" SWEET SPOT
            if prob[2] >= 0.40:       # X-Class trigger: 30% confidence
                batch_predictions.append(2)
                
            elif prob[1] >= 0.35:     # M-Class trigger: 35% confidence
                batch_predictions.append(1)
                
            else:
                batch_predictions.append(0) # Otherwise, it's Safe
                
        predicted = torch.tensor(batch_predictions)
        all_preds.extend(predicted.cpu().numpy())
        all_actuals.extend([get_class_index(t.item()) for t in targets])

pred_labels = [classes[i] for i in all_preds]
actual_labels = [classes[i] for i in all_actuals]

print("\n📊 --- RECALL CLASSIFICATION REPORT ---")
print(classification_report(actual_labels, pred_labels, labels=classes, zero_division=0))

plt.figure(figsize=(7, 5))
cm = confusion_matrix(actual_labels, pred_labels, labels=classes)

sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', 
            xticklabels=classes, yticklabels=classes, 
            annot_kws={"size": 14, "weight": "bold"})

plt.title('Threat Detection Matrix: Focus on M-Class Recall', fontweight='bold')
plt.xlabel('AI Prediction', fontweight='bold')
plt.ylabel('Actual Event', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import os
from PIL import Image
import torch
from torch.utils.data import Dataset
from torchvision import transforms

CSV_PATH = '/kaggle/input/datasets/organizations/fhnw-i4ds/sdobenchmark/SDOBenchmark_example/training/meta_data.csv'
IMAGE_FOLDER_PATH = '/kaggle/input/datasets/organizations/fhnw-i4ds/sdobenchmark/SDOBenchmark_example/training/'

print("🔍 Parsing the exact Kaggle folder paths...")

df_full = pd.read_csv(CSV_PATH)
valid_data = []

for index, row in df_full.iterrows():
    folder_id = str(row['id'])  # e.g., "12391_2015_08_02_23_34_00_0"
    
    # THE FIX: Split the ID into exactly two parts!
    # parts[0] becomes "12391"
    # parts[1] becomes "2015_08_02_23_34_00_0"
    parts = folder_id.split('_', 1) 
    
    if len(parts) == 2:
        ar_number = parts[0]
        timestamp_folder = parts[1]
        # Build the exact path: .../training/12391/2015_08_02_23_34_00_0
        event_folder = os.path.join(IMAGE_FOLDER_PATH, ar_number, timestamp_folder)
    else:
        event_folder = os.path.join(IMAGE_FOLDER_PATH, folder_id)

    actual_path = None
    if os.path.isdir(event_folder):
        files = os.listdir(event_folder)
        
        # We specifically want the Magnetogram images
        mag_files = [f for f in files if 'magnetogram' in f.lower() and f.lower().endswith(('.jpg', '.png'))]
        img_files = [f for f in files if f.lower().endswith(('.jpg', '.png'))] 
        
        if mag_files:
            actual_path = os.path.join(event_folder, mag_files[0])
        elif img_files: # Fallback to any image if magnetogram is missing
            actual_path = os.path.join(event_folder, img_files[0])
            
    if actual_path:
        valid_data.append({
            'id': folder_id,
            'peak_flux': row['peak_flux'],
            'file_path': actual_path
        })

clean_df = pd.DataFrame(valid_data)
print(f"✅ PURGE COMPLETE: Successfully mapped {len(clean_df)} images out of {len(df_full)} metadata rows.")

class SolarFlareCleanDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.data = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path = self.data.iloc[idx]['file_path']
        image = Image.open(img_path).convert('L') 
        target_flux = float(self.data.iloc[idx]['peak_flux'])
        if self.transform:
            image = self.transform(image)
        return image, torch.tensor(target_flux, dtype=torch.float32)

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

clean_test_dataset = SolarFlareCleanDataset(dataframe=clean_df, transform=transform)
```

```python
import gradio as gr
import random
import torch
import torch.nn.functional as F
import numpy as np

# 1. Classes
ui_classes = ['Safe (B/C)', 'M-Class', 'X-Class']

def get_magnitude_str(flux):
    if flux >= 1e-4: return f"X{flux/1e-4:.2f}"
    if flux >= 1e-5: return f"M{flux/1e-5:.2f}"
    if flux >= 1e-6: return f"C{flux/1e-6:.2f}"
    return f"B{flux/1e-7:.2f}"

device = next(model.parameters()).device
model.eval()

# ── History store (in-memory, last scan only) ─────────────────────────────────
history = {"img": None, "ai": "", "actual": "", "conf": "", "status": ""}

def build_conf_html(p_safe, p_m, p_x):
    return f"""
    <div style="display:flex; gap:10px;">
      <div style="flex:1; background:#1e2330; border:1px solid #2a3040; border-radius:10px; padding:14px;">
        <div style="font-size:11px; color:#6b7a99; text-transform:uppercase; letter-spacing:0.07em; margin-bottom:10px;">Safe (B/C)</div>
        <div style="height:5px; border-radius:3px; background:#2a3040; margin-bottom:10px;">
          <div style="height:5px; border-radius:3px; background:#4ade80; width:{p_safe}%;"></div>
        </div>
        <div style="font-size:22px; font-weight:600; color:#4ade80;">{p_safe}<span style="font-size:14px; font-weight:400; color:#6b7a99;">%</span></div>
      </div>
      <div style="flex:1; background:#1e2330; border:1px solid #2a3040; border-radius:10px; padding:14px;">
        <div style="font-size:11px; color:#6b7a99; text-transform:uppercase; letter-spacing:0.07em; margin-bottom:10px;">M-Class</div>
        <div style="height:5px; border-radius:3px; background:#2a3040; margin-bottom:10px;">
          <div style="height:5px; border-radius:3px; background:#f59e0b; width:{p_m}%;"></div>
        </div>
        <div style="font-size:22px; font-weight:600; color:#f59e0b;">{p_m}<span style="font-size:14px; font-weight:400; color:#6b7a99;">%</span></div>
      </div>
      <div style="flex:1; background:#1e2330; border:1px solid #2a3040; border-radius:10px; padding:14px;">
        <div style="font-size:11px; color:#6b7a99; text-transform:uppercase; letter-spacing:0.07em; margin-bottom:10px;">X-Class</div>
        <div style="height:5px; border-radius:3px; background:#2a3040; margin-bottom:10px;">
          <div style="height:5px; border-radius:3px; background:#f87171; width:{p_x}%;"></div>
        </div>
        <div style="font-size:22px; font-weight:600; color:#f87171;">{p_x}<span style="font-size:14px; font-weight:400; color:#6b7a99;">%</span></div>
      </div>
    </div>
    """

def build_ai_html(ai_category, ai_mag_str):
    return f"""
    <div style="background:#1e2330; border:1px solid #2a3040; border-radius:10px; padding:16px;">
      <div style="font-size:11px; color:#6b7a99; text-transform:uppercase; letter-spacing:0.07em; margin-bottom:8px;">🤖 AI Prediction</div>
      <div style="font-size:20px; font-weight:600; color:#e2e8f0; margin-bottom:4px;">{ai_category}</div>
      <div style="font-size:13px; color:#94a3b8;">Magnitude: <span style="color:#cbd5e1; font-weight:500;">{ai_mag_str}</span></div>
    </div>
    """

def build_actual_html(actual_category, actual_mag_str):
    return f"""
    <div style="background:#1e2330; border:1px solid #2a3040; border-radius:10px; padding:16px;">
      <div style="font-size:11px; color:#6b7a99; text-transform:uppercase; letter-spacing:0.07em; margin-bottom:8px;">🛰️ Actual Physics</div>
      <div style="font-size:20px; font-weight:600; color:#e2e8f0; margin-bottom:4px;">{actual_category}</div>
      <div style="font-size:13px; color:#94a3b8;">Magnitude: <span style="color:#cbd5e1; font-weight:500;">{actual_mag_str}</span></div>
    </div>
    """

def build_status_html(ai_category, actual_category, predicted_idx):
    if ai_category == actual_category:
        bg, border, color, icon, msg = "#0f2a1a", "#166534", "#4ade80", "✅", "Success — AI accurately caught the event."
    elif predicted_idx > ui_classes.index(actual_category):
        bg, border, color, icon, msg = "#2a1f0a", "#854d0e", "#f59e0b", "⚠️", "False alarm — AI overestimated the threat level."
    else:
        bg, border, color, icon, msg = "#2a0f0f", "#991b1b", "#f87171", "❌", "Critical miss — system failed to identify the major flare."
    return f"""
    <div style="background:{bg}; border:1px solid {border}; border-radius:10px; padding:14px 18px; display:flex; align-items:center; gap:12px;">
      <span style="font-size:18px;">{icon}</span>
      <div>
        <div style="font-size:14px; font-weight:600; color:{color};">{msg}</div>
        <div style="font-size:12px; color:#6b7a99; margin-top:3px;">AI: {ai_category} &nbsp;·&nbsp; Actual: {actual_category}</div>
      </div>
    </div>
    """

# ── Previous: just replay history ─────────────────────────────────────────────
def show_previous():
    if history["img"] is None:
        no_prev = "<div style='font-size:13px; color:#4a5568; font-style:italic; padding:8px 0;'>No previous scan available yet.</div>"
        return None, no_prev, no_prev, no_prev, no_prev
    return (
        history["img"],
        history["ai"],
        history["actual"],
        history["conf"],
        history["status"],
    )

# ── Main scan ─────────────────────────────────────────────────────────────────
def fetch_and_predict_ui():
    max_retries = 15
    for attempt in range(max_retries):
        try:
            idx = random.randint(0, len(test_dataset) - 1)
            test_img_tensor, actual_flux_tensor = test_dataset[idx]

            img_array = test_img_tensor.squeeze().cpu().numpy()
            display_img = ((img_array - img_array.min()) / (img_array.max() - img_array.min()) * 255).astype(np.uint8)

            input_tensor = test_img_tensor.unsqueeze(0).to(device)
            actual_flux = actual_flux_tensor.item()

            with torch.no_grad():
                logits, flux_log_pred = model(input_tensor)
                probabilities = F.softmax(logits, dim=1).squeeze().cpu().numpy()
                predicted_idx = int(np.argmax(probabilities))
                predicted_flux = 10 ** flux_log_pred.item()

            ai_category    = ui_classes[predicted_idx]
            ai_mag_str     = get_magnitude_str(predicted_flux)

            if actual_flux >= 1e-4:   actual_category = 'X-Class'
            elif actual_flux >= 1e-5: actual_category = 'M-Class'
            else:                     actual_category = 'Safe (B/C)'

            actual_mag_str = get_magnitude_str(actual_flux)

            p_safe = int(round(float(probabilities[0]) * 100))
            p_m    = int(round(float(probabilities[1]) * 100))
            p_x    = 100 - p_safe - p_m

            ai_html     = build_ai_html(ai_category, ai_mag_str)
            actual_html = build_actual_html(actual_category, actual_mag_str)
            conf_html   = build_conf_html(p_safe, p_m, p_x)
            status_html = build_status_html(ai_category, actual_category, predicted_idx)

            # Save current as history BEFORE returning new results
            history["img"]    = display_img
            history["ai"]     = ai_html
            history["actual"] = actual_html
            history["conf"]   = conf_html
            history["status"] = status_html

            return display_img, ai_html, actual_html, conf_html, status_html

        except FileNotFoundError:
            continue
        except Exception as e:
            err = f"""<div style="background:#2a0f0f; border:1px solid #991b1b; border-radius:10px; padding:14px; color:#f87171; font-size:14px;"><b>System error:</b> {str(e)}</div>"""
            return None, "", "", "", err

    err = """<div style="background:#2a0f0f; border:1px solid #991b1b; border-radius:10px; padding:14px; color:#f87171; font-size:14px;"><b>System error:</b> Dataset paths are broken.</div>"""
    return None, "", "", "", err


# ── DARK CSS ──────────────────────────────────────────────────────────────────
dark_css = """
body, .gradio-container, .main, footer { background:#0d1117 !important; }
.gr-panel, .gr-box, .gr-form, [class*="block"] { background:transparent !important; border:none !important; box-shadow:none !important; }

#sg-header {
    display:flex; align-items:center; gap:14px;
    padding:20px 24px 18px;
    border-bottom:1px solid #1e2330;
    margin-bottom:20px;
}
#sg-logo {
    width:42px; height:42px; border-radius:50%;
    background:#1a2540; border:1px solid #2a3a60;
    display:flex; align-items:center; justify-content:center; font-size:20px; flex-shrink:0;
}
#sg-title  { font-size:20px; font-weight:700; color:#e2e8f0; margin:0; }
#sg-sub    { font-size:13px; color:#6b7a99; margin-top:2px; }
#sg-live   { margin-left:auto; display:flex; align-items:center; gap:7px; }
#sg-dot    { width:9px; height:9px; border-radius:50%; background:#4ade80; box-shadow:0 0 6px #4ade8088; }
#sg-livetxt{ font-size:12px; font-weight:600; color:#4ade80; }

.sg-card {
    background:#161b27 !important;
    border:1px solid #1e2330 !important;
    border-radius:14px !important;
    padding:20px !important;
    margin:0 8px !important;
}
.sg-label {
    font-size:11px !important; font-weight:600 !important;
    color:#6b7a99 !important; text-transform:uppercase !important;
    letter-spacing:0.08em !important; margin-bottom:14px !important;
    display:block !important;
}
#feed-img { border-radius:10px !important; background:#1e2330 !important; }
#feed-img img { border-radius:10px !important; }

/* Scan button */
#scan-btn button {
    background: linear-gradient(135deg, #1d4ed8, #2563eb) !important;
    color: #ffffff !important;
    border: 1px solid #3b82f6 !important;
    border-radius: 10px !important;
    font-size: 14px !important;
    font-weight: 600 !important;
    padding: 12px !important;
    width: 100% !important;
    margin-top: 14px !important;
    cursor: pointer !important;
    letter-spacing: 0.03em !important;
    transition: all 0.2s !important;
}
#scan-btn button:hover { background: linear-gradient(135deg, #1e40af, #1d4ed8) !important; transform: translateY(-1px) !important; }
#scan-btn button:active { transform: translateY(0) scale(0.98) !important; }

/* Previous button */
#prev-btn button {
    background: transparent !important;
    color: #6b7a99 !important;
    border: 1px solid #2a3040 !important;
    border-radius: 10px !important;
    font-size: 13px !important;
    font-weight: 500 !important;
    padding: 10px !important;
    width: 100% !important;
    margin-top: 8px !important;
    cursor: pointer !important;
    letter-spacing: 0.02em !important;
    transition: all 0.2s !important;
}
#prev-btn button:hover {
    background: #1e2330 !important;
    color: #94a3b8 !important;
    border-color: #3a4560 !important;
}
#prev-btn button:active { transform: scale(0.98) !important; }

.sg-divider { height:1px; background:#1e2330; margin:16px 0; }
.sg-standby { font-size:13px; color:#4a5568; font-style:italic; padding:8px 0; }
.prose, .md, [data-testid="markdown"] { background: transparent !important; color: #e2e8f0 !important; }
"""

# ── LAYOUT ────────────────────────────────────────────────────────────────────
with gr.Blocks(css=dark_css, theme=gr.themes.Base()) as dashboard:

    gr.HTML("""
    <div id="sg-header">
      <div id="sg-logo">🛰️</div>
      <div>
        <div id="sg-title">SolarGuard</div>
        <div id="sg-sub">Automated orbital watch · SDO magnetogram analysis</div>
      </div>
      <div id="sg-live">
        <div id="sg-dot"></div>
        <span id="sg-livetxt">Live</span>
      </div>
    </div>
    """)

    with gr.Row():

        # ── LEFT: Feed ──────────────────────────────────────────────────────
        with gr.Column(scale=4, elem_classes=["sg-card"]):
            gr.HTML("<span class='sg-label'>📡 SDO magnetogram feed</span>")
            feed_display = gr.Image(
                interactive=False,
                image_mode="L",
                height=280,
                show_label=False,
                container=False,
                elem_id="feed-img",
            )
            scan_btn = gr.Button("↺  Scan next active region", elem_id="scan-btn")
            prev_btn = gr.Button("← Previous result",          elem_id="prev-btn")

        # ── RIGHT: Analysis ─────────────────────────────────────────────────
        with gr.Column(scale=6, elem_classes=["sg-card"]):
            gr.HTML("<span class='sg-label'>🔬 Post-event analysis — AI vs physics</span>")

            with gr.Row():
                ai_output     = gr.HTML("<div class='sg-standby'>Waiting for scan...</div>")
                actual_output = gr.HTML("<div class='sg-standby'>Waiting for scan...</div>")

            gr.HTML("<div class='sg-divider'></div>")
            gr.HTML("<span class='sg-label'>Class confidence</span>")
            conf_output = gr.HTML("<div class='sg-standby'>Run a scan to see confidence scores.</div>")

            gr.HTML("<div class='sg-divider'></div>")
            status_output = gr.HTML("<div class='sg-standby'>System standing by...</div>")

    # ── Wire up ───────────────────────────────────────────────────────────────
    outputs = [feed_display, ai_output, actual_output, conf_output, status_output]

    scan_btn.click(fn=fetch_and_predict_ui, inputs=[], outputs=outputs)
    prev_btn.click(fn=show_previous,        inputs=[], outputs=outputs)

dashboard.launch(share=True, inline=True)
```
